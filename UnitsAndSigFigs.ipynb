```python
def scientific_notation_10(x):
    xx=x.split('e')
    sup=''
    #print(int(xx[1]))
    if int(xx[1])==0:
        return xx[0]+' '
    else:
        for i in range(len(xx[1])):
            sup+=unicode_super(xx[1][i])
        return xx[0]+' '+u"\u00D7"+'10'+sup
    
def round_to_sig_fig(num, unit='', sigfig=3, scientific_notation=False, suffix='', prefix=''):
    if type(unit)==str:
        symbol=unit_symbol(unit)
        divide = multiplier(unit)
    elif type(unit)==float:
        divide=unit
        symbol=''
    
    ans_temp=num/divide
    
    digits_before_decimal=0
    zeros_after_decimal=0
    
    while (ans_temp>10**digits_before_decimal):
        digits_before_decimal+=1
    
    while (ans_temp<10**(zeros_after_decimal-1)):
        zeros_after_decimal-=1
    
    round_upto=sigfig-digits_before_decimal-zeros_after_decimal
    ans=round(ans_temp, round_upto)
    
    if scientific_notation==True:
        form='{:.'+str(sigfig-1)+'e}'
        #(sig-1) is there because python forces the result to be in the format x.xx
        temp=form.format(ans)
        return scientific_notation_10(temp)+prefix+symbol+suffix
    elif scientific_notation==False:
        if round_upto>0:
            form='{:.'+str(round_upto)+'f}'
        elif round_upto<=0:
            form='{:.0f}'
        return form.format(ans)+' '+prefix+symbol+suffix
    
def unit_symbol(u):
    prefix = {'deci': 'd',
             'centi': 'c',
             'mili': 'm',
             'micro': '\u03BC',
             'nano': 'n',
             'pico': 'p',
             'femto': 'f',
             'atto': 'a',
             'zepto': 'z',
             'yocto': 'y',
             
             'deca': 'da',
             'hecto': 'h',
             'kilo': 'k',
             'mega': 'M',
             'giga': 'G',
             'tera': 'T',
             'peta': 'P',
             'exa': 'E',
             'zetta': 'Y',
             'yotta': 'Z',
             
             '': ''}
    return prefix[u]

def multiplier(u):
    multiply = {'deci': 1e-1,
             'centi': 1e-2,
             'mili': 1e-3,
             'micro': 1e-6,
             'nano': 1e-9,
             'pico': 1e-12,
             'femto': 1e-15,
             'atto': 1e-18,
             'zepto': 1e-21,
             'yocto': 1e-24,
             
             'deca': 1e+1,
             'hecto': 1e+2,
             'kilo': 1e+3,
             'mega': 1e+6,
             'giga': 1e+9,
             'tera': 1e+12,
             'peta': 1e+15,
             'exa': 1e+18,
             'zetta': 1e+21,
             'yotta': 1e+24,
             
             '': 1}
    return multiply[u]

def unicode_super(u):
    if type(u)==int:
        u=str(u)
    sup = {'0':u"\u2070",
          '1':u"\u00B9",
          '2':u"\u00B2",
          '3':u"\u00B3",
          '4':u"\u2074",
          '5':u"\u2075",
          '6':u"\u2076",
          '7':u"\u2077",
          '8':u"\u2078",
          '9':u"\u2079",
          '-':u"\u207B",
          '+':''
          }
    return sup[u]


temp1=2.56500147e-02
temp2=2.67620467e-18
temp3=10.4352435e-07

print(round_to_sig_fig(temp2, unit='', sigfig=3, scientific_notation=True, suffix=''))
```
